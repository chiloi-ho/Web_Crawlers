```python
title_list = []
content_list = []
count = 0

for url in url_list:
    res = requests.get(url)
    
    if "/2018/" in url or "/2019/" in url or "/2020/" in url or "2021" in url:
        res.encoding = 'utf-8' 
    elif "/2016/" in url or "/2017/" in url:
        res.encoding = 'GBK'
    else:
        res.encoding = "gb2312"
        
    soup = BeautifulSoup(res.text, 'html.parser')
    
    if res.encoding == "gb2312":
        title = soup.find("title")
    else:
        title = soup.find('h1') 
        
    try:
        title = title.text.strip()
        print(count, title)
        title_list.append(title)
    except Exception:
        title = "403 Forbidden"
        title_list.append(title)
    
    news_contents = ""
    if "shipin" in url:
        contents = ""
    elif title == "403 Forbidden":
        contents = ""
    else:
        try:
            contents = soup.find('div', 'left_zw').find_all('p')
        except Exception:
            contents = ""
    for content in contents:
        if 'function' in content.text:
            continue
        news_contents = news_contents + content.text.strip()
    content_list.append(news_contents)
    
    count += 1
    
print("---------------ALL DONE---------------")
```
